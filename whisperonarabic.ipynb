```python
from transformers import pipeline
```

```python
pipe = pipeline(model="Huzaifatahir/whisper-small-ar")
```

```python
def transcribe(audio):
    text = pipe(audio)["text"]
    text_ar = text
    print(text_ar)
    return text_ar

audio = "/content/drive/MyDrive/ColabNotebooks/arabic_medical_words_audio_1/word_13.mp3"

transcribe(audio)
```
